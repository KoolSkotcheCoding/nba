```python
import pandas as pd
import numpy as np
import json
import math
from decimal import Decimal
import collections
from operator import itemgetter

%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns

from IPython.display import IFrame

sns.set_color_codes()
sns.set_style("white")
```

```python
with open('/nbaData/0021500491.json') as data_file:
    data=json.load(data_file)
```

```python
#moments 
#contains (quarter, ?, game clock (seconds), shot clock (seconds),
#          ?, player+ball position===position)

#shot clock can be empty (none, nonetype, etc.)
#position contains: (team ID, player ID, x position, y position, z position)

#For ball, team ID = -1, player ID = -1, z position could be physically meaningful
#For players, z position is always zero. 

class NBA_moment:
    def __init__(self,rawMoment,event):
        self.event=event
        self.quarter=rawMoment[0]
        if rawMoment:
            self.gameClock=rawMoment[2]
            self.shotClock=rawMoment[3]
        else:
            self.gameClock=float(-1)
            self.shotClock=float(-1)
        
```

```python
class NBA_event:
    def __init__(self,rawEvent):
        self.eventID=int(rawEvent["eventId"])
        self.moments=[NBA_moment(moment,self.eventID) for moment in rawEvent["moments"]]

    
```

```python
class NBA_game:
    def __init__(self, rawGame):
        self.events=[NBA_event(event) for event in rawGame["events"]]
        self.gameDate=rawGame["gamedate"]
        self.gameID=rawGame["gameid"]
        
        #Create a list of 
        self.chron_moments=[]
        for eventIndex, event in enumerate(self.events):
            for momentIndex, moment in enumerate(event.moments):
                self.chron_moments.append([(eventIndex,momentIndex),moment.gameClock])
        self.chron_moments=sorted(self.chron_moments, key=itemgetter(1), reverse=True)
        
```

```python
testGame=NBA_game(data)

print testGame.gameID
print testGame.gameDate
#print testGame.events[0].moments[22].event
#print testGame.events[0].moments[22].shotClock
#print testGame.chron_moments
```

```python
#The below plots are evidence that the moments within a given event
#can overlap with potentially many other event's moments.

#It does not appear that any moments within an event overlap 
#with other moments in that same event. 


Q1GameClock=[]
Q1ShotClock=[]
for event in testGame.events:
    #print type(event.eventID)
    if event.eventID<3:
        
        for moment in event.moments:
        #rint moment.quarter
            if moment.quarter==1:
            #print moment.quarter
                Q1GameClock.append(moment.gameClock/60)
                Q1ShotClock.append(moment.shotClock)
                
#game clock plot
plt.scatter(range(0,len(Q1GameClock)),Q1GameClock)
plt.title('game clocks in moments of consecutive events')
plt.ylabel('Game Time Left (min)')
plt.xlabel('moment')
plt.show()

#shot clock plot
plt.scatter(range(0,len(Q1ShotClock)),Q1ShotClock)
plt.title('shot clocks in moments of consecutive events')
plt.ylabel('Shot Clock Time Left (s)')
plt.xlabel('moment')
axes=plt.gca()
axes.set_xlim([0,len(Q1ShotClock)])
plt.show()

#chron check
breakCount=0
breakID=[]
for event in testGame.events:
    PrevClock=12*60
    for moment in event.moments:
        if (12*60-moment.gameClock)<0:
            print "somethings wrong here"
        elif (PrevClock-moment.gameClock)<0:
            breakCount+=1
            breakID.append(event.eventID)
        PrevClock=moment.gameClock

if breakCount:
    print "There are ",breakCount," event-level discontinuities within the moments of the game"
#print breakID
```
